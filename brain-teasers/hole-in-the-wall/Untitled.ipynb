```python
import random
import numpy.random

def sim(y):
    right = 1
    location = np.random.triangular(-right, 0, right)
    y = right * y
    
    to_right = location > 0
    location = abs(location)
    
    if not(to_right):
        return 2 * y + location
    else:
        if location < y:
            return location
        else:
            return 2 * y + 2 * right + location

def simN(y, N):
    return np.mean([ sim(y) for _ in range(N) ])

simN(0.66, 50000)
```

```python
step = 0.05
xs = np.arange(0, 1 + step, step)
ys = [ simN(x, 500000) for x in xs ]
plt.plot(xs, ys)
```
